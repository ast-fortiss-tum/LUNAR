## Test Execution

We execute LUNAR to test ConvNavi as follows:

```python
#!/bin/bash

mkdir -p results/simple_run/run

PYTHONUNBUFFERED=1 LLM_IPA="gpt-5-chat" python run_mt_navi.py \
    --algorithm rs \
    --n 15 \
    --sut openai \
    --llm_ipa gpt-5-chat \
    --llm_intent_classifier DeepSeek-V3-0324 \
    --llm_judge gpt-5-mini \
    --llm_generator DeepSeek-V3-0324 \
    --features_config configs/features_simple_judge_cc.json \
    --max_time "00:05:00" \
    --store_turns_details \
    --weight_clarity 0.35 \
    --weight_request_orientedness 0.65 \
    --th_dims 0.65 \
    --th_efficiency 0.65 \
    --th_effectiveness 0.75 \
    --save_folder "results/simple_run/run" \
    --no_wandb
```
