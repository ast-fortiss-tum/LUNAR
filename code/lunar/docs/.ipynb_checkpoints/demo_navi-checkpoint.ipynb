## Test Execution

Execute LUNAR to test ConvNavi as follows:

```python
#!/bin/bash

mkdir -p results/simple_run/run

PYTHONUNBUFFERED=1 LLM_IPA="gpt-5-chat" python run_mt_navi.py \
    --algorithm rs \
    --n 15 \
    --sut openai \
    --llm_ipa gpt-5-chat \                       # THE LLM WE WANT TO TEST
    --llm_intent_classifier DeepSeek-V3-0324 \           # THE LLM USED TO CLASSIFY INTENTS OF ASSISTANT RESPONSES
    --llm_judge gpt-5-mini \                             # THE LLM THAT EVALUATES LINGUISTIC PROPERTIES OF RESPONSES
    --llm_generator DeepSeek-V3-0324 \                   # THE LLM THAT GENERATED UTTERANCES
    --features_config configs/features_simple_judge_cc.json \                      # THE SEARCH SPACE DEFINITION
    --max_time "00:05:00" \                              # MAXIMAL RUNTIME
    --store_turns_details \          
    --weight_clarity 0.35 \                              # WEIGHTS DEFINED FOR EVALUATION      
    --weight_request_orientedness 0.65 \         
    --th_dims 0.65 \                                     # THRESHOLD FOR JUDGE SCORE  
    --th_efficiency 0.65 \                               # THRESHOLD FOR EFFICIENCY SCORE
    --th_effectiveness 0.75 \                            # THRESHOLD FOR EFFECTIVENESS SCORE
    --save_folder "results/simple_run/run" \
    --no_wandb                                           # WANDB LOGGING DISABLED
```
